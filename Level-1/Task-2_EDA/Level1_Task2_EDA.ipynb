```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set default plot style and size
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12
sns.set_style('whitegrid')

print("Libraries loaded successfully!")
```

```python
df = pd.read_csv('/content/drive/MyDrive/churn-bigml-80 coddeva/churn_cleaned coddeva level 1 task1.csv')

print("Dataset loaded successfully!")
print("Number of rows    :", df.shape[0])
print("Number of columns :", df.shape[1])
df.head(10)
```

```python
print("=" * 50)
print("GENERAL DATASET INFORMATION")
print("=" * 50)
df.info()
```

```python
print("=" * 50)
print("STATISTICAL SUMMARY")
print("=" * 50)
df.describe().round(2)
```

```python
num_cols = df.select_dtypes(include='number').columns

print("=" * 50)
print("MEAN")
print("=" * 50)
print(df[num_cols].mean().round(2))

print("\n" + "=" * 50)
print("MEDIAN")
print("=" * 50)
print(df[num_cols].median().round(2))

print("\n" + "=" * 50)
print("MODE (Most Frequent Value)")
print("=" * 50)
print(df[num_cols].mode().iloc[0])

print("\n" + "=" * 50)
print("STANDARD DEVIATION")
print("=" * 50)
print(df[num_cols].std().round(2))
```

```python
plt.figure(figsize=(7, 5))
churn_counts = df['Churn'].value_counts()
colors = ['steelblue', 'coral']

bars = plt.bar(['Stays (0)', 'Churns (1)'],
               churn_counts.values,
               color=colors,
               edgecolor='black',
               width=0.5)

# Add count labels on top of bars
for bar, val in zip(bars, churn_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 20,
             str(val),
             ha='center',
             fontsize=13,
             fontweight='bold')

plt.title('Churn Distribution', fontsize=15, fontweight='bold')
plt.xlabel('Customer Status')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.savefig('churn_distribution.png', dpi=150)
plt.show()
print(" Chart saved: churn_distribution.png")
```

```python
num_cols = df.select_dtypes(include='number').columns.tolist()
num_cols.remove('Area code')  # Not useful for analysis

fig, axes = plt.subplots(4, 5, figsize=(20, 14))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    axes[i].hist(df[col], bins=25, color='steelblue',
                 edgecolor='black', alpha=0.8)
    axes[i].set_title(col, fontsize=10, fontweight='bold')
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Count')

# Hide empty subplots
for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

plt.suptitle('Distribution of All Numerical Columns (Histograms)',
             fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('histograms.png', dpi=150)
plt.show()
print("Chart saved: histograms.png")
```

```python
box_cols = ['Total day minutes', 'Total eve minutes',
            'Total night minutes', 'Total intl minutes',
            'Customer service calls', 'Account length',
            'Total day charge', 'Total intl calls']

plt.figure(figsize=(16, 7))
df[box_cols].boxplot(patch_artist=True,
                     boxprops=dict(facecolor='lightblue'),
                     medianprops=dict(color='red', linewidth=2))

plt.title('Boxplot — Outlier Detection',
          fontsize=15, fontweight='bold')
plt.xticks(rotation=30, ha='right')
plt.ylabel('Value')
plt.tight_layout()
plt.savefig('boxplot.png', dpi=150)
plt.show()
print(" Chart saved: boxplot.png")
```

```python
plt.figure(figsize=(10, 6))

# Color by Churn status
colors = df['Churn'].map({0: 'steelblue', 1: 'coral'})

plt.scatter(df['Total day minutes'],
            df['Total day charge'],
            c=colors, alpha=0.5, edgecolors='none', s=40)

# Legend
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='steelblue', label='Stays (0)'),
    Patch(facecolor='coral', label='Churns (1)')
]
plt.legend(handles=legend_elements, fontsize=11)

plt.title('Total Day Minutes vs Total Day Charge',
          fontsize=15, fontweight='bold')
plt.xlabel('Total Day Minutes')
plt.ylabel('Total Day Charge')
plt.tight_layout()
plt.savefig('scatter.png', dpi=150)
plt.show()
print("Chart saved: scatter.png")
```

```python
plt.figure(figsize=(16, 12))

corr_matrix = df.corr(numeric_only=True)

sns.heatmap(corr_matrix,
            annot=True,
            fmt='.2f',
            cmap='coolwarm',
            linewidths=0.5,
            linecolor='white',
            square=True,
            annot_kws={'size': 9})

plt.title('Correlation Matrix',
          fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('heatmap.png', dpi=150)
plt.show()
print(" Chart saved: heatmap.png")
```

```python
compare_cols = ['Total day minutes', 'Customer service calls',
                'Account length', 'Total intl calls']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, col in enumerate(compare_cols):
    sns.boxplot(x='Churn', y=col, data=df,
                hue='Churn',
                palette={0: 'steelblue', 1: 'coral'},
                legend=False,
                ax=axes[i])
    axes[i].set_title(f'{col} by Churn Status',
                      fontsize=12, fontweight='bold')
    axes[i].set_xlabel('Churn (0 = Stays, 1 = Churns)')
    axes[i].set_ylabel(col)

plt.suptitle('Comparative Analysis by Churn Status',
             fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('churn_comparison.png', dpi=150)
plt.show()
print(" Chart saved: churn_comparison.png")
```

```python
print("=" * 60)
print("         FINAL FINDINGS — EDA SUMMARY")
print("=" * 60)

total = len(df)
churned = df['Churn'].sum()
not_churned = total - churned

print(f"\n Total customers             : {total}")
print(f" Retained customers          : {not_churned} ({not_churned/total*100:.1f}%)")
print(f" Churned customers           : {churned} ({churned/total*100:.1f}%)")

print(f"\n Avg. total day minutes      : {df['Total day minutes'].mean():.1f}")
print(f" Avg. total eve minutes      : {df['Total eve minutes'].mean():.1f}")
print(f" Avg. total night minutes    : {df['Total night minutes'].mean():.1f}")
print(f" Avg. customer service calls : {df['Customer service calls'].mean():.2f}")
print(f" Avg. account length         : {df['Account length'].mean():.1f} days")

print(f"\n Top 5 Correlated Feature Pairs:")
corr = df.corr(numeric_only=True).abs()
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
top_corr = upper.stack().sort_values(ascending=False).head(5)
print(top_corr.round(2))

print("\n Key Observations:")
print(f"   - {churned/total*100:.1f}% of customers churned")
print(f"   - Customers with high day minutes tend to churn more")
print(f"   - Higher customer service calls linked to higher churn risk")
print(f"   - Total minutes and total charge are highly correlated")

print("\n EDA completed successfully!")
```
