```python
import pandas as pd
import numpy as np

print(" Libraries loaded successfully!")
```

```python
df = pd.read_csv('/content/drive/MyDrive/codveda level 1 task 1 2 3/churn-bigml-80 coddeva.csv')
print(" Dataset loaded successfully!")
print("Number of rows    :", df.shape[0])
print("Number of columns :", df.shape[1])
df.head(10)
```

```python
print("=" * 50)
print("GENERAL DATASET INFORMATION")
print("=" * 50)
df.info()
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
# Fill numerical columns with mean
for col in df.select_dtypes(include='number').columns:
    df[col].fillna(df[col].mean(), inplace=True)

# Fill categorical columns with mode
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

print(" Missing values handled. Verification:")
print(df.isnull().sum())
```

```python
print("Duplicate rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print(" Duplicates removed. New shape:", df.shape)
```

```python
# Convert True/False and Yes/No to 1/0
df['Churn'] = df['Churn'].map({True: 1, False: 0})
df['International plan'] = df['International plan'].map({'Yes': 1, 'No': 0})
df['Voice mail plan'] = df['Voice mail plan'].map({'Yes': 1, 'No': 0})

print(" Categorical values converted:")
df[['Churn', 'International plan', 'Voice mail plan']].head()
```

```python
df.to_csv('churn_cleaned.csv', index=False)
print("File saved: churn_cleaned.csv")
```
